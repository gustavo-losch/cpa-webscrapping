```python
from flask import Flask, request, jsonify
import pandas as pd
```

```python
data = pd.read_csv('Games_Data.csv', encoding='latin', low_memory=False)
data.head()
```

```python
data = data.fillna(value={'release_date': '-', 'developer': '-', 'publisher': '-',
                   'genres': '-', 'multiplayer_or_singleplayer': '-'}) #colocando '-' em valores faltantes
data = data.drop(columns='id')
data.head()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data = data.dropna()
data.describe()
```

```python
print(data['win_support'].unique())
data = data[data['win_support'] != '79%'] #removendo uma linha onde win_support era '79%'
print(data['win_support'].unique())
```

```python
data = data.astype({'mac_support': 'int32', 'lin_support': 'int32'}) #mudando numberos binarios para int
data['percent_positive'] = data['percent_positive'].str.rstrip('%').astype(int) #mudando percent(string) para int
data.head()
```

```python
data.to_csv('clean_data.csv', index=False)
```
