```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset 1
df_penguins = pd.read_csv("penguins.csv")

# Dataset 2
df_etapa_teorica = pd.DataFrame()
df_etapa_teorica["f1"] = [0.02, -2.0, 0.57, 1.0, -2.0, 0.84, 0.66, -1.0, 0.44, 0.32]
df_etapa_teorica["f2"] = [True, False, True, False, True, True, False, False, True, True]
df_etapa_teorica["f3"] = ["red", "red", "yellow", "black", "black", "red", "red", "yellow", "red", "red"]
df_etapa_teorica["f4"] = [8.25, 1.35, 21.86, 3.06, 1.58, 1.25, 1.68, 1.01, 1.19, 1.47]
```

## I) Etapa Teórica

1) Quais  atributos  são  categóricos  e  quais  atributos  são  numéricos? 
Apenas  com  as  informações  dadas  é  possível  identificar  dentre  os 
atributos numéricos quais são os tipos? 
 
2) Qual a moda de f3: 
 
3) Compute a média, mediana, desvio padrão, variância, dos atributos 
f1 e f4. 
 
4) Calcule a covariância entre os atributos f1 e f4. 
 
5) Calcule a correlação existente entre os atributos f1 e f4

```python
# 1
df_etapa_teorica.info()
'''
1. Categóricos (categorical): categories or groups = f2, f3
2. Numéricos (numerical): measurable quantities = f1, f4
'''
```

```python
# 2
mode = df_etapa_teorica["f3"].mode() # mode is a series
print(f'Valor mais frequente: {mode.iloc[0]}')
```

```python
# 3
f1 = np.array(df_etapa_teorica["f1"])
f4 = np.array(df_etapa_teorica["f4"])

mean_f1, mean_f4, median_f1, median_f4 = df_etapa_teorica["f1"].mean(), df_etapa_teorica["f2"].mean(), df_etapa_teorica["f1"].median(), df_etapa_teorica["f2"].median()
standard_deviation_f1, standard_deviation_f4 = f1.std(), f4.std()
variance_f1, variance_f4 = f1.var(), f4.var()

print(
    "Média-F1:", mean_f1, 
    "\nMédia-F4:", mean_f4, 
    "\nMediana-F1:", median_f1,
    "\nMediana-F4:", median_f4, 
    "\nDesvioPadrão-F1:", standard_deviation_f1, 
    "\nDesvioPadrão-F4:", standard_deviation_f4,
    "\nVariância-F1:", variance_f1,
    "\nVariância-F4:", variance_f4
)
```

```python
# 4
covariance = df_etapa_teorica["f1"].cov(df_etapa_teorica["f4"])
print(covariance)
```

```python
# 5
correlation = df_etapa_teorica["f1"].corr(df_etapa_teorica["f4"])
print(correlation) # 0.25 (-1, 1)
```

```python
# 2
# Checking if there are any missing values:
missing_values = df_penguins.isnull().any()
print(missing_values)
```

```python
missing_values = df_penguins.isnull().sum()
total_missing_values = df_penguins.isnull().sum().sum()
print(missing_values)
print(f'Total: {total_missing_values}')
```

```python
# 3
unique_species = df_penguins["Species"].unique()
for species in unique_species:
    count = df_penguins["Species"].value_counts()[species]
    print(f'{species}: {count}')
print(f'Total: {df_penguins["Species"].count()}')
```

```python
# Em português para facilitar
# Criar uma figura com dois subplots lado a lado
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Configuração do primeiro gráfico (Boxplot)
sns.boxplot(x="Island", 
            y="Culmen Length (mm)", 
            hue="Species",
            palette="Set2",
            width=0.6,
            data=df_penguins, 
            ax=ax1)

# Personalização do primeiro gráfico
ax1.set_title("Boxplot do Comprimento do Culmen por Ilha", fontsize=14)
ax1.set_xlabel("Ilha", fontsize=12)
ax1.set_ylabel("Comprimento do Culmen (mm)", fontsize=12)
ax1.legend(title="Espécie")

# Configuração do segundo gráfico (Scatter plot)
sns.scatterplot(x="Culmen Length (mm)", 
                y="Culmen Depth (mm)", 
                hue="Species", 
                palette="colorblind",
                s=100,
                data=df_penguins,
                ax=ax2)

# Personalização do primeiro gráfico
ax2.set_title("Relação entre Comprimento e Profundidade do Culmen", fontsize=14)
ax2.set_xlabel("Comprimento do Culmen (mm)", fontsize=12)
ax2.set_ylabel("Profundidade do Culmen (mm)", fontsize=12)
ax2.legend(title="Espécie", fontsize=10, title_fontsize=12)

# Ajuste do layout
plt.tight_layout()
plt.show()
```

### Explicação dos gráficos
##### _Tópicos_

1) 
- O Boxplot mostra a distribuição do comprimento do culmen em cada ilha, separado por espécies.
- Não tem todas as espécies em todas as ilhas.
- A distribuição individual das espécies são semelhantes independente das ilhas, sugerindo que as caracteristicas morfológicas são genéticas e não moldadas pelo ambiente.

2) 
- O Scatterplot mostra a separacão morfolófica entre as espécies.
- Adelie: culmen mais curto e profundo.
- Gentoo: culmen longo e pouco profundo.
- Chinstrap: culmen mais longo e profundidade média.
